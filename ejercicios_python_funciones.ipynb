Problema 1: Producto de Matrices

Implemente una función llamada mat_mul que calcule el resultado del producto de dos matrices
A y B. La función debe contar con una mínima documentación y debe testear que A y B tengan
dimensiones compatibles. El script tiene que llamarse mat_mul.py



```python
def mat_mul(A, B):

    # Verificacion de dimensiones
    if len(A) != len(B[0]):
        
        return print("Las matrices no son compatibles para multiplicar.") 
    else:
    
    # creación de la dimension de la matriz resultado
    
        C_f=[]
        C=[]

        for x in range(0,len(A[0]),1):
            C_f.append(0)
        
        for x in range(0,len(B),1):
            C.append(C_f[:])


        # Cálculo del producto
        for i in range(0, len(A[0]),1): 
            for j in range(0, len(B),1):
                for r in range(0, len(A),1):
                    C[j][i]= C[j][i] + (A[r][i] * B[j][r])
                    
        return C
```

Problema 2: Factorial

El factorial de un número entero no negativo n, denotado como n!, se define como el producto de
todos los enteros positivos menores o iguales a n,
n! = n × (n − 1) × (n − 2) × . . . × 1, con la convención de que 0! = 1.
Implemente dos versiones del cálculo del factorial: (i) la versión recursiva (que usa invocaciones a la misma función) y (ii) la iterativa (que usa un for para el cálculo).

```python
def factorial(lista,metodo='iterativo'):
    """Este script tiene dos metodos para resolver el factorial de una lista de números, 
    devuelve una lista con los factoriales
    este mismo tendrá como defecto la opción iterativo pero se puede cambiar a recursivo según preferencia
    metodo =iteractivo - metodo=recursivo"""

    if   metodo=='iterativo':
        #se realizará un for por que conocemos la cantidad de veces que debemos iterar
        m=len(lista)
        lista_resultado_factoriales=[]
        for x in range(0,m,1):
            lista_resultado_factoriales.append(1) 
            if lista[x]>0:
                for y in range(0,lista[x],1):
                     lista_resultado_factoriales[x] *= (lista[x]-y)

            elif lista[x]<0:
               raise ZeroDivisionError(f"No se admiten números negativos en la lista verificar el valor: {lista[x]} esta en el index:{x}")
        return lista_resultado_factoriales
        
    elif metodo=='recursivo':
        
        def factorial_recursivo(n):
            if n ==0:
                return 1
            return n * factorial_recursivo(n-1)
            
        m=len(lista)
        lista_resultado_factoriales=[]

        for x in range(0,m,1):
            lista_resultado_factoriales.append(1)
            if lista[x]>0:
                lista_resultado_factoriales[x]=(factorial_recursivo(lista[x])) 
            elif lista[x]<0:
               raise ZeroDivisionError(f"No se admiten números negativos en la lista verificar el valor: {lista[x]} esta en el index:{x}")
        return lista_resultado_factoriales       

    
```
